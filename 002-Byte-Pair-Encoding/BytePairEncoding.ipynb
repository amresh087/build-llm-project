# Byte Pair Encoding

<div class="alert alert-block alert-info">
    If we want implements BPE functionally starting taking more time. There are a libery call tiktoken which is used chart gpt.  https://github.com/openai/tiktoken
</div>


```python
import importlib
import tiktoken
print("tiktoken version", importlib.metadata.version("tiktoken"))
```

```python
tokenizer=tiktoken.get_encoding("gpt2")
```

```python
text=("Hello, do you like tea? <|endoftext|> In the sunlite terraces of someunknownPlace.")
intergers=tokenizer.encode(text, allowed_special={"<|endoftext|>"})
print(intergers)
```

<div class="alert alert-block alert-info">
    text convert into token ID
</div>

<div class="alert alert-block alert-info">
    Now we can convert token Id to ID
</div>

```python
strings=tokenizer.decode(intergers)
print(strings)
```
