# Creating-Input-output-pair

```python
import tiktoken
with open("the-verdict.txt","r", encoding="utf-8")as f:
    raw_text = f.read()

print("total number of charecter", len(raw_text))
    
print(raw_text[:100])        
```

```python
tokenizer=tiktoken.get_encoding("gpt2")

enc_text=tokenizer.encode(raw_text)
print(len(enc_text))
```
